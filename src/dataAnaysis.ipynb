```python
import pandas as pd
import numpy as np

# Assuming your DataFrame is named df

# Define a function to extract CB_email and Department_email
def extract_emails(value):
    if isinstance(value, dict):
        cb_emails = value.get('CB_email', set())
        dept_emails = value.get('Department_email', set())
        return cb_emails, dept_emails
    else:
        return set(), set()

# Apply the function to create the 'CB_email' and 'Department_email' columns
df[['CB_email', 'Department_email']] = df['Emails'].apply(lambda x: pd.Series(extract_emails(x)))

# Display the DataFrame with the new columns
print(df)
```

```python
import pandas as pd
import numpy as np
import ast  # Library for literal string parsing

# Assuming your DataFrame is named df

# Define a function to parse the string representation of the dictionary
def parse_emails(value):
    if isinstance(value, str):
        # Parse the string to a dictionary
        email_dict = ast.literal_eval(value)
        cb_emails = email_dict.get('CB_email', set())
        dept_emails = email_dict.get('Department_email', set())
        return cb_emails, dept_emails
    else:
        return set(), set()

# Apply the function to create the 'CB_email' and 'Department_email' columns
df[['CB_email', 'Department_email']] = df['Emails'].apply(lambda x: pd.Series(parse_emails(x)))

# Display the DataFrame with the new columns
print(df)
```

```python
# Function to remove newline characters
import re
def remove_newline(phone_numbers):
    if isinstance(phone_numbers, str):
        return re.sub(r'\n', '', phone_numbers)
    else:
        return phone_numbers

# Apply the function to remove newline characters
df['Phone Numbers'] = df['Phone Numbers'].apply(remove_newline)
```
